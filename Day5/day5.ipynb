```python
with open('d5_stack.txt') as f:
    boxes = [line.rstrip('\n').replace('[',' ').replace(']',' ') for line in f]
boxes
```

```python
stack_height = len(boxes)-1
print(stack_height)
```

```python
stack_list=[]
for row in boxes:
    stack_row=[]
    for i in range(1, len(row)-1, 4):
        stack_row.append(row[i])
    stack_list.append(stack_row)
print(stack_list)
```

```python
test_dict = {}
for column in zip(*stack_list): 
    t_list = []
    new_list = list(column[0:len(column)-1])
    for l in list(reversed(new_list)):
        if l != ' ':
            t_list.append(l)
    test_dict[column[len(column)-1]] = t_list
# test_dict
```

```python
test_dict
```

```python
with open('d5_move.txt') as f:
    moves = [line.rstrip('\n').split(' ') for line in f]
moves
```

```python
popping_dict = test_dict.copy()
for row in moves:
    box_no = int(row[1])
    start_stack = row[3]
    end_stack = row[5]
    start_len = len(popping_dict[start_stack])
#     b_to_m = popping_dict[start_stack][-box_no:start_len]
    b_to_m = popping_dict[start_stack][-box_no:]

    popping_dict[end_stack].extend(b_to_m)

    for item in b_to_m:
        popping_dict[start_stack].reverse()
        popping_dict[start_stack].remove(item)
        popping_dict[start_stack].reverse()

    # Part 1
#     for i in range(0, box_no, 1):
#         removed_box = popping_dict[start_stack].pop()
#         popping_dict[end_stack].append(removed_box)


print(popping_dict)
```

```python
final_string = ''
for key in popping_dict:
    final_string += popping_dict[key][-1]
print(final_string)
```

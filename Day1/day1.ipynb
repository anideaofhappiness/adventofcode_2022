```python
with open('d1.txt') as f:
    calories = [line.rstrip('\n') for line in f]
calories
```

```python
def add_calories(calories):
    cal_add = []
    temp_sum = 0
    for line in calories:
#         print(line)
        if not line:
            cal_add.append(temp_sum)
            temp_sum = 0
        else:
            temp_sum = temp_sum + int(line)  
    return cal_add
```

```python
elf_calories = add_calories(calories)
```

```python
sorted(elf_calories, reverse=True)
```

```python
sum(sorted(elf_calories, reverse=True)[0:3])
```

```python
def sum_top_calories(cal_list, no_elves):
    sum_top_cal = sum(sorted(cal_list, reverse=True)[0:no_elves])
    return sum_top_cal
```

```python
sum_top_calories(elf_calories, 1)
```

```python
sum_top_calories(elf_calories, 3)
```
